```python
import pandas as pd
import numpy as np
import time
data = './data/'
#显示所有列
pd.set_option('display.max_columns', None)
#显示所有行
pd.set_option('display.max_rows', None)
```

```python
train = pd.read_csv(data + 'round1_iflyad_anticheat_traindata.txt', sep='\t', encoding='utf-8')
print(train.shape)
print(train.info())
train.head()
```

```python
#count_list = ['label','apptype','nginxtime_date','city','province']
#train[['ip']].head()
count_list = ['dvctype']
for cl in count_list:
    count = train[[cl]].groupby(cl).size()
    print(count)
```

```python

train[['nginxtime']].head()
train['nginxtime_format'] = train.apply(lambda x: time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(x['nginxtime']/1000)), axis=1)
train['nginxtime_date'] = train.apply(lambda x: x['nginxtime_format'].split(' ')[0], axis=1)
train['nginxtime_time'] = train.apply(lambda x: x['nginxtime_format'].split(' ')[1], axis=1)
train['nginxtime_hour'] = train.apply(lambda x: x['nginxtime_format'].split(' ')[1].split(':')[0], axis=1)
train.head()
```

```python
test = pd.read_csv(data + 'round1_iflyad_anticheat_testdata_feature.txt', sep='\t', encoding='utf-8')
test['nginxtime_format'] = test.apply(lambda x: time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(x['nginxtime']/1000)), axis=1)
test['nginxtime_date'] = test.apply(lambda x: x['nginxtime_format'].split(' ')[0], axis=1)
print(test.shape)
print(test.groupby('nginxtime_date').size())
test.head()
```
